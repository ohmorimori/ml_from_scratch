```python


import sys
dir_str = ".."
if (dir_str not in sys.path):
        sys.path.append(dir_str)

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from utils import mean_squared_error
from supervised_learning import RegressionTree
```

```python
data = pd.read_csv("../data/TempLinkoping2016.txt", sep="\t")
time = np.array(data["time"].values).reshape(-1, 1)
temp = np.array(data["temp"].values).reshape(-1, 1)

X =  (time - np.min(time)) / (np.max(time) - np.min(time))
y = temp[:, 0]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RegressionTree()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

#color map
cmap = plt.get_cmap('viridis')
mse = mean_squared_error(y_test, y_pred)

print("Mean Squared Error: ", mse)
#plot results
m1 = plt.scatter(336 * X_train, y_train, color=cmap(0.9), s=10)
m2 = plt.scatter(336 * X_test, y_test, color=cmap(0.5), s=10)
m3 = plt.scatter(336 * X_test, y_pred, color='black', s=10)

plt.suptitle("Regression Tree")
plt.title("MSE: %.2f" % mse, fontsize=10)
plt.xlabel('Day')
plt.ylabel('Temperature in Celcius')
plt.legend((m1, m2, m3), ("Training data", "Test data", "Prediction"), loc="lower right")
plt.show()
```
