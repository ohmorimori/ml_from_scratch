```python
import sys
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

dir_str = ".."
if dir_str not in sys.path:
    sys.path.append(dir_str)

from utils import train_test_split, normalize, accuracy_score, euclidean_distance, Plot
from supervised_learning import Knn
```

```python
data = datasets.load_iris()
X = normalize(data.data)
y = data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
clf = Knn(k=5)
y_pred = clf.predict(X_test, X_train, y_train)

accuracy = accuracy_score(y_test, y_pred)
print ("Accuracy:", accuracy)

# Reduce dimensions to 2d using pca and plot the results
Plot().plot_in_2d(X_test, y_pred, title="K Nearest Neighbors", accuracy=accuracy, legend_labels=data.target_names)
```
