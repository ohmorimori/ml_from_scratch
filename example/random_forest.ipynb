```python
import sys
dir_str = ".."
if (dir_str not in sys.path):
        sys.path.append(dir_str)
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from utils import accuracy_score, Plot
from supervised_learning import RandomForest
```

```python
data = datasets.load_digits()
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)

X_train.shape
```

```python
clf = RandomForest(n_estimators=100)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: ", accuracy)
plt = Plot()
plt.plot_in_2d(X_test, y_pred, title="Random Forest", accuracy=accuracy, legend_labels=data.target_names)
```
